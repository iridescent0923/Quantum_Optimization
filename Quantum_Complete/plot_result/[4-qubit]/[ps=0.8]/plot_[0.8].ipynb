```python
import matplotlib.pyplot as plt
from IPython.display import display, Latex
from matplotlib.ticker import MultipleLocator
import numpy as np  # Original numpy

from enum import Enum
```

```python
data_files = {
    'data_0': 'result_[0]_new3.npy',
    'data_0.005': 'result_[0.005]_new.npy',
    'data_0.05': 'result_[0.05]_new.npy',
    'data_0.1': 'result_[0.1]_new.npy',
    'data_0.5': 'result_[0.5].npy'
}

data = {}
for key, file_name in data_files.items():
    data[key] = np.load(file_name)
```

```python
class INDEX(Enum):
    PHI = 0
    CFI = 1
    
    THETA_X1 = 2
    THETA_X2 = 3
    
    PHI_Z1 = 4
    PHI_Z2 = 5
    PHI_Z3 = 6
    PHI_Z4 = 7
    
    TAU_L1 = 8
    TAU_L2 = 9
    
    TAU_R1 = 10
    TAU_R2 = 11

class DataPlotter:
    def __init__(self, data_in):
        self.data_in = data_in
        
    def plot_cfi(self, tau_dephase=0, gamma_ps=0):        
        plt.plot(
            self.data_in[0][:,INDEX.PHI.value], 
            self.data_in[0][:,INDEX.CFI.value], 
            label = f'$\\tau$ = {tau_dephase}')
        
        plt.title(f'CFI at $\gamma_{{ps}} = {gamma_ps}$')
        plt.xlabel('Time')
        plt.ylabel('CFI')
        plt.grid()
        plt.legend()
        # plt.show()
        
    def plot_theta_x1(self, tau_dephase=0, set_ylim=(0, 0)):
        plt.plot(
            self.data_in[0][:,INDEX.PHI.value], 
            self.data_in[0][:,INDEX.THETA_X1.value], 
            label = f'$\\tau$ = {tau_dephase}')
        
        self.plot_settings(
            title = f'Optimized $\\theta_{{x1}}$',
            ylim = set_ylim
        )

    def plot_theta_x2(self, tau_dephase=0, set_ylim=(0, 0)):
        plt.plot(
            self.data_in[0][:,INDEX.PHI.value], 
            self.data_in[0][:,INDEX.THETA_X2.value], 
            label = f'$\\tau$ = {tau_dephase}')
        
        self.plot_settings(
            title = f'Optimized $\\theta_{{x2}}$',
            ylim = set_ylim
        )
        
    def plot_phi_z1(self, tau_dephase=0, set_ylim=(0, 0)):
        plt.plot(
            self.data_in[0][:,INDEX.PHI.value], 
            self.data_in[0][:,INDEX.PHI_Z1.value], 
            label = f'$\\tau$ = {tau_dephase}')
        
        self.plot_settings(
                    title = f'Optimized $\\phi_{{z1}}$',
                    ylim = set_ylim
        )
        
    def plot_phi_z2(self, tau_dephase=0, set_ylim=(0, 0)):
        plt.plot(
            self.data_in[0][:,INDEX.PHI.value], 
            self.data_in[0][:,INDEX.PHI_Z2.value], 
            label = f'$\\tau$ = {tau_dephase}')
        
        self.plot_settings(
                    title = f'Optimized $\\phi_{{z2}}$',
                    ylim = set_ylim
        )
        
    def plot_phi_z3(self, tau_dephase=0, set_ylim=(0, 0)):
        plt.plot(
            self.data_in[0][:,INDEX.PHI.value], 
            self.data_in[0][:,INDEX.PHI_Z3.value], 
            label = f'$\\tau$ = {tau_dephase}')

        self.plot_settings(
                    title = f'Optimized $\\phi_{{z3}}$',
                    ylim = set_ylim
        )
        
    def plot_phi_z4(self, tau_dephase=0, set_ylim=(0, 0)):
        plt.plot(
            self.data_in[0][:,INDEX.PHI.value], 
            self.data_in[0][:,INDEX.PHI_Z4.value], 
            label = f'$\\tau$ = {tau_dephase}')

        self.plot_settings(
                    title = f'Optimized $\\phi_{{z4}}$',
                    ylim = set_ylim
        )
        
    def plot_tau_L1(self, tau_dephase=0, set_ylim=(0, 0)):
        plt.plot(
            self.data_in[0][:,INDEX.PHI.value], 
            self.data_in[0][:,INDEX.TAU_L1.value], 
            label = f'$\\tau$ = {tau_dephase}')
        
        self.plot_settings(
                    title = f'Optimized $\\tau_{{L1}}$',
                    ylim = set_ylim
        )
        
    def plot_tau_L2(self, tau_dephase=0, set_ylim=(0, 0)):
        plt.plot(
            self.data_in[0][:,INDEX.PHI.value], 
            self.data_in[0][:,INDEX.TAU_L2.value], 
            label = f'$\\tau$ = {tau_dephase}')

        self.plot_settings(
                    title = f'Optimized $\\tau_{{L2}}$',
                    ylim = set_ylim
        )
        
    def plot_tau_R1(self, tau_dephase=0, set_ylim=(0, 0)):
        plt.plot(
            self.data_in[0][:,INDEX.PHI.value], 
            self.data_in[0][:,INDEX.TAU_R1.value], 
            label = f'$\\tau$ = {tau_dephase}')

        self.plot_settings(
                    title = f'Optimized $\\tau_{{R1}}$',
                    ylim = set_ylim
        )
        
    def plot_tau_R2(self, tau_dephase=0, set_ylim=(0, 0)):
        plt.plot(
            self.data_in[0][:,INDEX.PHI.value], 
            self.data_in[0][:,INDEX.TAU_R2.value], 
            label = f'$\\tau$ = {tau_dephase}')

        self.plot_settings(
                    title = f'Optimized $\\tau_{{R2}}$',
                    ylim = set_ylim
        )
        
    def plot_settings(self, title, xlabel='Time', ylabel='Rad', ylim=(-2*np.pi, 2*np.pi)):
        plt.yticks(
            [-np.pi, -3*np.pi/4, -np.pi/2, -np.pi/4, 0, np.pi/4, np.pi/2, 3*np.pi/4, np.pi, (3*np.pi)/2, 2*np.pi, 5*np.pi/2], 
            ['$-\pi$', '$-3\pi/4$', '$-\pi/2$', '$-\pi/4$', '0', '$\pi/4$', '$\pi/2$', '$3\pi/4$', '$\pi$', '$3\pi/2$', '$2\pi$', '$5\pi/2$']
        )
        plt.ylim(ylim)
        plt.title(title)
        plt.xlabel(xlabel)
        plt.ylabel(ylabel)
        plt.grid()
        plt.legend()
        # plt.show()
```
