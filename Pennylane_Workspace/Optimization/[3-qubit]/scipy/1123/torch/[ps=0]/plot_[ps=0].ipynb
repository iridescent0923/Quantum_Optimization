```python
import matplotlib.pyplot as plt
from IPython.display import display, Latex
from matplotlib.ticker import MultipleLocator
import numpy as np  # Original numpy

import pandas as pd
from enum import Enum
```

```python
data_0 = pd.read_csv('[0].csv').values
data_1 = pd.read_csv('[0.005].csv').values
data_2 = pd.read_csv('[0.05].csv').values
data_3 = pd.read_csv('[0.5].csv').values
data_4 = pd.read_csv('[0.1].csv').values

FULL_DATA = np.zeros((
    5,
    len(data_0[:, 0]),
    len(data_0[0, :])
))

# FULL_DATA[0][:, :]

FULL_DATA[0][:, :] = data_0
FULL_DATA[1][:, :] = data_1
FULL_DATA[2][:, :] = data_2
FULL_DATA[3][:, :] = data_3
FULL_DATA[4][:, :] = data_4
```

```python
class INDEX(Enum):
    PHI = 0
    CFI = 1
    
    THETA_X = 2
    
    PHI_Z1 = 3
    PHI_Z2 = 4
    PHI_Z3 = 5
    
    TAU_1 = 6
    TAU_2 = 7
```

```python
tau_list = 0, 0.005, 0.05, 0.5, 0.1
for tau_idx, tau_current in enumerate(tau_list):
    plt.plot(FULL_DATA[tau_idx][:,INDEX.PHI.value], FULL_DATA[tau_idx][:,INDEX.CFI.value], label = f'$\\tau$ = {tau_current}')
plt.title(f'CFI at $\gamma_{{ps}} = {0}$')
plt.xlabel('Time')
plt.ylabel('CFI')
plt.grid()
plt.legend()
plt.show()
```

```python
for tau_idx, tau_current in enumerate(tau_list):
    plt.plot(FULL_DATA[tau_idx][:,INDEX.PHI.value], FULL_DATA[tau_idx][:,INDEX.THETA_X.value], label = f'$\\tau$ = {tau_current}')

plt.yticks(
    [-np.pi, -np.pi/2, 0, np.pi/2, np.pi, (3*np.pi)/2, 2*np.pi], 
    ['$-\pi$', '$-\pi/2$', '0', '$\pi/2$', '$\pi$', '$3\pi/2$', '$2\pi$']
)
plt.ylim(-np.pi/2, np.pi/2)
plt.title(f'Optimized $\\theta_{{x}}$')
plt.xlabel('Time')
plt.ylabel('RAD')
plt.grid()
plt.legend()
plt.show()
```

```python
for tau_idx, tau_current in enumerate(tau_list):
    plt.plot(FULL_DATA[tau_idx][:,INDEX.PHI.value], FULL_DATA[tau_idx][:,INDEX.PHI_Z1.value], label = f'$\\tau$ = {tau_current}')

plt.yticks(
    [-np.pi, -np.pi/2, 0, np.pi/2, np.pi, (3*np.pi)/2, 2*np.pi, 5*np.pi/2, 3*np.pi], 
    ['$-\pi$', '$-\pi/2$', '0', '$\pi/2$', '$\pi$', '$3\pi/2$', '$2\pi$', '$5\pi/2$', '$3\pi$']
)

plt.ylim(0, 4*np.pi/2)
plt.title(f'Optimized $\\phi_{{z1}}$')
plt.xlabel('Time')
plt.ylabel('RAD')
plt.grid()
plt.legend()
plt.show()
```
