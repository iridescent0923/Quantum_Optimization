```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from IPython.display import display, Latex
from matplotlib.ticker import MultipleLocator
```

```python
data_0 = np.load('[0].npy')                 # 3d arr
data_1 = np.load('[0.05].npy')              # data_1[0][:, 0] = 5e-2
data_2 = pd.read_csv('[0.5].csv').values    # 5e-1
data_2_PHI = np.arange(1e-2, 6.282, 1e-3)
```

```python
tau_list = 0, 5e-3, 5e-2, 1e-1, 5e-1, 1
idx_add = 0
for tau_idx, current_tau in enumerate(tau_list):
    if current_tau == 5e-2:
        plt.plot(data_1[0][:, 0], data_1[0][:, 1+idx_add], label = f'$\\tau$ = {current_tau}')
    elif current_tau == 5e-1:
        plt.plot(data_2_PHI, data_2[:, 0+idx_add], label = f'$\\tau$ = {current_tau}')
    else:
        plt.plot(data_0[tau_idx][:, 0+idx_add], data_0[tau_idx][:, 1], label = f'$\\tau$ = {current_tau}')
    # 0 5e-3 5e-2 1e-1 5e-1 1
plt.title(f'CFI at $\gamma_{{ps}} = {0.8}$')
plt.xlabel('Time')
plt.ylabel('RAD')
plt.grid()
plt.legend()
plt.show()
```

```python
tau_list = 0, 5e-3, 5e-2, 1e-1, 5e-1, 1
idx_add = 1
for tau_idx, current_tau in enumerate(tau_list):
    if current_tau == 5e-2:
        plt.plot(data_1[0][:, 0], data_1[0][:, 1+idx_add], label = f'$\\tau$ = {current_tau}')
    elif current_tau == 5e-1:
        plt.plot(data_2_PHI, data_2[:, 0+idx_add], label = f'$\\tau$ = {current_tau}')
    else:
        plt.plot(data_0[tau_idx][:, 0+idx_add], data_0[tau_idx][:, 1], label = f'$\\tau$ = {current_tau}')
    # 0 5e-3 5e-2 1e-1 5e-1 1
plt.yticks(
    [-np.pi, -np.pi/2, 0, np.pi/2, np.pi, (3*np.pi)/2, 2*np.pi], 
    ['$-\pi$', '$-\pi/2$', '0', '$\pi/2$', '$\pi$', '$3\pi/2$', '$2\pi$']
)
plt.ylim(0, 1*np.pi)
plt.title(f'Optimized $\\theta_{{x}}$')
plt.xlabel('Time')
plt.ylabel('RAD')
plt.grid()
plt.legend()
plt.show()
```
