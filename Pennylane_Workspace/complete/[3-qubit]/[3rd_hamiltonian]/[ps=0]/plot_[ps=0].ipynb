```python
import matplotlib.pyplot as plt
from IPython.display import display, Latex
from matplotlib.ticker import MultipleLocator
import numpy as np  # Original numpy

import pandas as pd
from enum import Enum
```

```python
data_0 = pd.read_csv('[0_new].csv').values
data_1 = pd.read_csv('[0.005].csv').values
data_2 = pd.read_csv('[0.05].csv').values
data_3 = pd.read_csv('[0.1].csv').values
data_4 = pd.read_csv('[0.5].csv').values
```

```python
class INDEX(Enum):
    PHI = 0
    CFI = 1
    
    THETA_X = 2
    
    PHI_Z1 = 3
    PHI_Z2 = 4
    PHI_Z3 = 5
    
    TAU_1 = 6
    TAU_2 = 7
```

```python
plt.plot(data_0[:,INDEX.PHI.value], data_0[:,INDEX.CFI.value], label = f'$\\tau$ = {0}')
plt.plot(data_1[:,INDEX.PHI.value], data_1[:,INDEX.CFI.value], label = f'$\\tau$ = {0.005}')
plt.plot(data_2[:,INDEX.PHI.value], data_2[:,INDEX.CFI.value], label = f'$\\tau$ = {0.05}')
plt.plot(data_3[:,INDEX.PHI.value], data_3[:,INDEX.CFI.value], label = f'$\\tau$ = {0.1}')
plt.plot(data_4[:,INDEX.PHI.value], data_4[:,INDEX.CFI.value], label = f'$\\tau$ = {0.5}')
plt.title(f'CFI at $\gamma_{{ps}} = {0}$')
plt.xlabel('Time')
plt.ylabel('CFI')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot(data_0[:,INDEX.PHI.value], data_0[:,INDEX.THETA_X.value], label = f'$\\tau$ = {0}')
plt.plot(data_1[:,INDEX.PHI.value], data_1[:,INDEX.THETA_X.value], label = f'$\\tau$ = {0.005}')
plt.plot(data_2[:,INDEX.PHI.value], data_2[:,INDEX.THETA_X.value], label = f'$\\tau$ = {0.05}')
plt.plot(data_3[:,INDEX.PHI.value], data_3[:,INDEX.THETA_X.value], label = f'$\\tau$ = {0.1}')
plt.plot(data_4[:,INDEX.PHI.value], data_4[:,INDEX.THETA_X.value], label = f'$\\tau$ = {0.5}')

plt.yticks(
    [-np.pi, -np.pi/2, 0, np.pi/2, np.pi, (3*np.pi)/2, 2*np.pi], 
    ['$-\pi$', '$-\pi/2$', '0', '$\pi/2$', '$\pi$', '$3\pi/2$', '$2\pi$']
)
plt.ylim(-np.pi/2, 2*np.pi/2)
plt.title(f'Optimized $\\theta_{{x}}$')
plt.xlabel('Time')
plt.ylabel('RAD')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot(data_0[:,INDEX.PHI.value], data_0[:,INDEX.TAU_1.value], label = f'$\\tau$ = {0}')
plt.plot(data_1[:,INDEX.PHI.value], data_1[:,INDEX.TAU_1.value], label = f'$\\tau$ = {0.005}')
plt.plot(data_2[:,INDEX.PHI.value], data_2[:,INDEX.TAU_1.value], label = f'$\\tau$ = {0.05}')
plt.plot(data_3[:,INDEX.PHI.value], data_3[:,INDEX.TAU_1.value], label = f'$\\tau$ = {0.1}')
plt.plot(data_4[:,INDEX.PHI.value], data_4[:,INDEX.TAU_1.value], label = f'$\\tau$ = {0.5}')

plt.yticks(
    [-np.pi, -np.pi/2, 0, np.pi/2, np.pi, (3*np.pi)/2, 2*np.pi], 
    ['$-\pi$', '$-\pi/2$', '0', '$\pi/2$', '$\pi$', '$3\pi/2$', '$2\pi$']
)
plt.ylim(0, 3*np.pi/4)
plt.title(f'Optimized $\\tau_{{1}}$')
plt.xlabel('Time')
plt.ylabel('RAD')
plt.grid()
plt.legend()
plt.show()
```
